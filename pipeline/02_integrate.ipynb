### [Helper Function]

```python
def load_file(path, **kwargs):
    ext = os.path.splitext(path)[1].lower()

    try:
        if ext == ".csv":
            try:
                return pd.read_csv(path, **kwargs)
            except Exception:
                try:
                    return pd.read_csv(path, sep=";", **kwargs)
                except Exception:
                    return pd.read_csv(
                        path,
                        engine="python",
                        on_bad_lines="skip",
                        **kwargs
                    )
        else:
            try:
                if ext == ".xls":
                    return pd.read_excel(path, engine="xlrd", **kwargs)
                return pd.read_excel(path, engine="openpyxl", **kwargs)
            except Exception:
                html_kwargs = {k: v for k, v in kwargs.items() if k in {"header", "skiprows", "index_col", "attrs", "match", "flavor", "displayed_only", "extract_links"}}
                return pd.read_html(path, **html_kwargs)[0]

    except Exception as e:
        raise Exception(
            f"Failed to read file:\n{path}\n\nError:\n{e}"
        )
    
def norm_puskesmas(s):
    if pd.isna(s):
        return np.nan
    return str(s).strip().upper()
```

## 1. Load Data Utama

```python
df_raw = load_file(RAW_FILE, sheet_name=0, header=None)
data_rows = df_raw.iloc[6:].reset_index(drop=True)
data_rows = data_rows[data_rows.iloc[:, 1].notna()].reset_index(drop=True)
print(f"Valid Rows: {len(data_rows)}")

identitas = pd.DataFrame()
for col_name, col_idx in IDENTITAS_COLS.items():
    identitas[col_name] = data_rows.iloc[:, col_idx]

def extract_bulan(data_rows, identitas, bulan_nama, col_map, is_detail=False):
    rows = []
    for i in range(len(data_rows)):
        row = identitas.iloc[i].to_dict()
        row["bulan"] = bulan_nama
        row["tahun"] = 2022
        for field, col_idx in col_map.items():
            if col_idx < data_rows.shape[1]:
                val = data_rows.iloc[i, col_idx]
                row[field] = val if pd.notna(val) else np.nan
            else:
                row[field] = np.nan
        if not is_detail:
            row["total_balita"] = np.nan
        rows.append(row)
    return pd.DataFrame(rows)

semua_bulan = []

for bulan, col_map in BULAN_JAN_MEI.items():
    df_bulan = extract_bulan(data_rows, identitas, bulan, col_map, is_detail=False)
    for col in ["lama_0_6","lama_6_11","lama_12_23","lama_24_59",
                "baru_0_6","baru_6_11","baru_12_23","baru_24_59",
                "lulus_membaik_0_6","lulus_membaik_6_11","lulus_membaik_12_23","lulus_membaik_24_59",
                "meninggal"]:
        df_bulan[col] = np.nan
    semua_bulan.append(df_bulan)

for bulan, col_map in BULAN_JUN_DES.items():
    df_bulan = extract_bulan(data_rows, identitas, bulan, col_map, is_detail=True)
    for col in ["u0_6","u6_11","u12_23","u24_59","kasus_baru","lulus","kasus_lama"]:
        if col not in df_bulan.columns:
            df_bulan[col] = np.nan
    semua_bulan.append(df_bulan)

df_main = pd.concat(semua_bulan, ignore_index=True)

urutan_bulan = ["Januari","Februari","Maret","April","Mei","Juni",
                "Juli","Agustus","September","Oktober","November","Desember"]
df_main["bulan"] = pd.Categorical(df_main["bulan"], categories=urutan_bulan, ordered=True)
df_main = df_main.sort_values(["kelurahan", "bulan"]).reset_index(drop=True)
df_main["puskesmas_key"] = df_main["puskesmas"].apply(norm_puskesmas)
print(f"Shape {df_main.shape}")
```

## 2. Load Data Pendukung

### a. Identitas Puskesmas

### f. Berat Bayi Lahir Rendah (BBLR)

```python
df_bblr = load_file(
    "data/raw/JUMLAH BAYI BERAT BADAN LAHIR RENDAH PER PUSKESMAS KOTA SEMARANG TAHUN 2022.xls"
    , header=1
)
df_bblr = df_bblr[df_bblr["Category"].notna()].copy()
df_bblr = df_bblr.rename(columns={"Category": "nama_puskesmas", "Kasus": "kasus_bblr"})
df_bblr["puskesmas_key"] = df_bblr["nama_puskesmas"].apply(norm_puskesmas)
print(f"  BBLR: {df_bblr.shape}")
```

### g. Ibu Hamil Risti

```python
df_risti = load_file(
    "data/raw/JUMLAH IBU HAMIL DAN RESIKO TINGGI PER PUSKESMAS KOTA SEMARANG TAHUN 2022.xls"
    , header=1
)
df_risti = df_risti[df_risti["Category"].notna()].copy()
df_risti = df_risti.rename(columns={
    "Category":    "nama_puskesmas",
    "hamil_risti": "hamil_risti",
    "ibu_hamil":   "ibu_hamil_total",
})
df_risti["puskesmas_key"] = df_risti["nama_puskesmas"].apply(norm_puskesmas)
df_risti["rasio_risti"] = df_risti["hamil_risti"] / df_risti["ibu_hamil_total"].replace(0, np.nan)
print(f"  Risti: {df_risti.shape}")
```

### h. Imunisasi per Bulan

```python
BULAN_IMUNISASI = [
    "Januari","Februari","Maret","April","Mei","Juni",
    "Juli","Agustus","September","Oktober","November"
]
imunisasi_list = []
for bulan in BULAN_IMUNISASI:
    if bulan == "Januari":
        path = f"data/raw/Penilaian Kinerja Puskesmas - Pelayanan Imunisasi per {bulan} 2022.xlsx"
    else:
        path = f"data/raw/Penilaian Kinerja Puskesmas Pkp Pelayanan Imunisasi {bulan} 2022.xls"
    if not os.path.exists(path):
        print(f"  [SKIP] {path}")
        continue
    if bulan == "Januari":
        df_tmp = load_file(path, header=1)
    else:
        df_tmp = load_file(path)
    if "Puskesmas" in df_tmp.columns:
        df_tmp = df_tmp[["Puskesmas", "Capaian"]].copy()
        df_tmp.columns = ["nama_puskesmas", "capaian_imunisasi"]
    else:
        df_tmp = df_tmp[["Category", "Pelayanan Imunisasi"]].copy()
        df_tmp.columns = ["nama_puskesmas", "capaian_imunisasi"]
    df_tmp = df_tmp[df_tmp["nama_puskesmas"].notna()].copy()
    df_tmp["bulan"] = bulan
    df_tmp["puskesmas_key"] = df_tmp["nama_puskesmas"].apply(norm_puskesmas)
    imunisasi_list.append(df_tmp)

df_imunisasi = pd.concat(imunisasi_list, ignore_index=True) if imunisasi_list else pd.DataFrame()
print(f"  Imunisasi bulanan: {df_imunisasi.shape}")
```

### i. Imunisasi Dasar Lengkap Kota

```python
df_idl = load_file(
    "data/raw/JUMLAH BAYI USIA 0 - 9 BULAN YANG IMUNISASI DASAR LENGKAP DI KOTA SEMARANG TAHUN 2022.xls"
)
df_idl = df_idl[df_idl["Category"].notna()].copy()
df_idl = df_idl.rename(columns={
    "Category":              "bulan",
    "Jumlah Laki-laki":      "idl_laki",
    "Jumlah Perempuan":      "idl_perempuan",
})
df_idl["idl_total"] = df_idl["idl_laki"] + df_idl["idl_perempuan"]

mapping_bulan = {
    "Jan": "Januari",
    "Feb": "Februari",
    "Mar": "Maret",
    "Apr": "April",
    "May": "Mei",
    "Mei": "Mei",
    "Jun": "Juni",
    "Jul": "Juli",
    "Aug": "Agustus",
    "Agu": "Agustus",
    "Ags": "Agustus",
    "Sep": "September",
    "Okt": "Oktober",
    "Nov": "November",
    "Des": "Desember",
}

df_idl["bulan"] = df_idl["bulan"].astype(str).str.strip().str[:3].map(mapping_bulan).fillna(df_idl["bulan"].astype(str).str.strip().str.capitalize())
print(f"  IDL kota: {df_idl.shape}")
```

## 3. Join Data

```python
df = df_main.copy()

# Join 1: Identitas puskesmas — per puskesmas
df = df.merge(
    df_identitas[["puskesmas_key","kode_puskesmas","latitude","longitude",
                  "luas_wilayah_km2","jumlah_penduduk","jumlah_kk","karakteristik_wilayah"]],
    on="puskesmas_key", how="left"
)

# Join 2: Sanitasi — per puskesmas
df = df.merge(
    df_sanitasi[["puskesmas_key","sanitasi_pilar1","sanitasi_pilar2",
                 "sanitasi_pilar3","sanitasi_pilar4","sanitasi_pilar5"]],
    on="puskesmas_key", how="left"
)

# Join 3: Ibu Hamil 4T — per puskesmas
df = df.merge(
    df_4t[["puskesmas_key","hamil_jarak2th","hamil_paritas",
           "hamil_umur_lebih35th","hamil_umur_kurang20th"]],
    on="puskesmas_key", how="left"
)

# Join 4: Kunjungan K4 — per puskesmas
df = df.merge(
    df_k4[["puskesmas_key","kunjungan_k4"]],
    on="puskesmas_key", how="left"
)

# Join 5: Kunjungan K1 — per puskesmas
df = df.merge(
    df_k1[["puskesmas_key","kunjungan_k1"]],
    on="puskesmas_key", how="left"
)

# Join 6: BBLR — per puskesmas
df = df.merge(
    df_bblr[["puskesmas_key","kasus_bblr"]],
    on="puskesmas_key", how="left"
)

# Join 7: Risti — per puskesmas
df = df.merge(
    df_risti[["puskesmas_key","hamil_risti","ibu_hamil_total","rasio_risti"]],
    on="puskesmas_key", how="left"
)

# Join 8: Imunisasi bulanan — per puskesmas
if not df_imunisasi.empty:
    df = df.merge(
        df_imunisasi[["puskesmas_key","bulan","capaian_imunisasi"]],
        on=["puskesmas_key","bulan"], how="left"
    )
else:
    df["capaian_imunisasi"] = np.nan

# Join 9: Imunisasi Dasar Lengkap kota — per bulan
df["bulan_str"] = df["bulan"].astype(str)
df_idl["bulan"] = df_idl["bulan"].astype(str)
df = df.merge(
    df_idl[["bulan","idl_laki","idl_perempuan","idl_total"]],
    left_on="bulan_str", right_on="bulan", how="left"
).drop(columns=["bulan_str", "bulan_y"]).rename(columns={"bulan_x": "bulan"})

df = df.drop(columns=["puskesmas_key"], errors="ignore")

df["bulan"] = pd.Categorical(df["bulan"], categories=urutan_bulan, ordered=True)
df = df.sort_values(["kelurahan","bulan"]).reset_index(drop=True)
```

```python
os.makedirs("data/integrated", exist_ok=True)
df.to_csv(OUTPUT_FILE, index=False)

print(f"\nOutput: {OUTPUT_FILE}")
```
