# Validate

## 0. Import

```python
import pandas as pd
import numpy as np
import os
from pathlib import Path

if not Path("data/integrated").exists() and Path("..").joinpath("data/integrated").exists():
    os.chdir("..")

df = pd.read_csv("data/integrated/integrated_data_stunting.csv")
```

## 1. Data Validation

### a. Summary

```python
print(f"Shape                  : {df.shape}")
print(f"Total Rows             : {len(df)}")
print(f"Total Columns          : {len(df.columns)}")

print(f"\nJumlah kecamatan       : {df['kecamatan'].nunique()}")
print(f"Jumlah kelurahan       : {df['kelurahan'].nunique()}")
print(f"Jumlah puskesmas       : {df['puskesmas'].nunique()}")
print(f"Rentang tahun          : {df['tahun'].min()} - {df['tahun'].max()}")
print(f"Daftar bulan           : {list(df['bulan'].dropna().unique())}")
print(f"Duplicate Rows         : {df.duplicated().sum()}")

print(f"\nData Types:\n{df.dtypes}")

print(f"\nData Sample (5 rows):\n{df.head()}")

print("\nDataset Columns:")
for i, col in enumerate(df.columns, 1):
    print(f"  {i:>3}. {col}")
```

### b. Missing Values

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(1)
missing_df = pd.DataFrame({"missing": missing, "pct": missing_pct})
missing_df = missing_df[missing_df["missing"] > 0].sort_values("missing", ascending=False)

if len(missing_df) == 0:
    print("Tidak ada missing value.")
else:
    print("\n-- Data utama --")
    utama = ["u0_6","u6_11","u12_23","u24_59","kasus_baru","kasus_lama","lulus",
             "lama_0_6","lama_6_11","lama_12_23","lama_24_59",
             "baru_0_6","baru_6_11","baru_12_23","baru_24_59","meninggal","total_balita"]
    for col in utama:
        if col in missing_df.index:
            print(f"  {col:<35}: {missing_df.loc[col,'missing']} ({missing_df.loc[col,'pct']}%)")

    print("\n-- Data pendukung --")
    pendukung = ["capaian_imunisasi","idl_total","idl_laki","idl_perempuan"]
    for col in pendukung:
        if col in missing_df.index:
            print(f"  {col:<35}: {missing_df.loc[col,'missing']} ({missing_df.loc[col,'pct']}%)")

    join_cols = ["sanitasi_pilar1","hamil_risti","kunjungan_k4","kunjungan_k1",
                 "kasus_bblr","latitude","capaian_imunisasi"]
    for col in join_cols:
        if col in missing_df.index:
            print(f"  {col:<35}: {missing_df.loc[col,'missing']} ({missing_df.loc[col,'pct']}%)")
```
