# Load Raw Data

## 0. Import

```python
import pandas as pd
import os
from pathlib import Path

if not Path("data/raw").exists() and Path("..").joinpath("data/raw").exists():
    os.chdir("..")
```

### [Helper Function]

```python
def load_file(path, **kwargs):
    ext = os.path.splitext(path)[1].lower()

    try:
        if ext == ".csv":
            try:
                return pd.read_csv(path, **kwargs)
            except Exception:
                try:
                    return pd.read_csv(path, sep=";", **kwargs)
                except Exception:
                    return pd.read_csv(
                        path,
                        engine="python",
                        on_bad_lines="skip",
                        **kwargs
                    )
        else:
            try:
                return pd.read_excel(path, **kwargs)
            except Exception:
                return pd.read_html(path, **kwargs)[0]

    except Exception as e:
        raise Exception(
            f"Failed to read file:\n{path}\n\nError:\n{e}"
        )
```

## 1. Dataset Utama

```python

files_utama = {
    "2022": "data/raw/622373002-TEPRA-STUNTING-THN-2022.xlsx",
}

dataframes_utama = {}
for tahun, path in files_utama.items():
    df = load_file(path)
    dataframes_utama[tahun] = df
    print(f"[OK] Stunting {tahun} — shape: {df.shape}")
```

## 2. Dataset Pendukung

### a. Imunisasi per Puskesmas per Bulan (11 file, Januari-November)

```python
BULAN_IMUNISASI = [
    "Januari","Februari","Maret","April","Mei","Juni",
    "Juli","Agustus","September","Oktober","November"
]
imunisasi_bulanan = {}
for bulan in BULAN_IMUNISASI:
    if bulan == "Januari":
        path = f"data/raw/Penilaian Kinerja Puskesmas - Pelayanan Imunisasi per {bulan} 2022.xlsx"
    else:
        path = f"data/raw/Penilaian Kinerja Puskesmas Pkp Pelayanan Imunisasi {bulan} 2022.xls"
    if os.path.exists(path):
        df = load_file(path)
        imunisasi_bulanan[bulan] = df
        print(f"[OK] Imunisasi {bulan} — shape: {df.shape}")
    else:
        print(f"[SKIP] File tidak ditemukan: {path}")
```
